```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://en.wikipedia.org/wiki/List_of_Nobel_laureates")
soup = BeautifulSoup(response.text, "html.parser")
table_tags = soup.find("table", attrs={"class":"wikitable sortable"})
row_tags = table_tags.find_all("td")

laureates=[]
for i in range(len(row_tags)) :
    laureates.append(row_tags[i].text.strip())
laureates.pop(6)
years = laureates[0::6]
Physicist = laureates[1::6]
Chemist = laureates[2::6]
physiologist = laureates[3::6]
literaturist = laureates[4::6]
peace = laureates[5::6]
econmic = "None"

years_1 = laureates[238::6]
Physicist_1 = laureates[239::6]
Chemist_1 = laureates[240::6]
physiologist_1 = laureates[241::6]
literaturist_1 = laureates[242::6]
peace_1 = laureates[243::6]

years_2 = laureates[394::7]
Physicist_2 = laureates[395::7]
Chemist_2 = laureates[396::7]
physiologist_2 = laureates[397::7]
literaturist_2 = laureates[398::7]
peace_2 = laureates[399::7]
econmic_2 = laureates[400::7]

with open("List_of_Nobel_laureates.csv", mode='w', encoding="utf-8") as fd:
    fd.write("year,physics,chemistry,physiology or medicine,literature,peace,econmic\n")
    for i in range(len(years)):
        if years[i]=="1940" :
            break
        fd.write(f"{years[i]},{Physicist[i]},{Chemist[i]},{physiologist[i]},{literaturist[i]},\"{peace[i]}\",{econmic}\n")
    fd.write(f"{1940},canclled,canclled,canclled,canclled,canclled,canclled\n")
    fd.write(f"{1941},canclled,canclled,canclled,canclled,canclled,canclled\n")
    fd.write(f"{1942},canclled,canclled,canclled,canclled,canclled,canclled\n")
    for j in range(len(years_1)):
        if years_1[j]=="1969" :
            break
        fd.write(f"{years_1[j]},{Physicist_1[j]},{Chemist_1[j]},{physiologist_1[j]},{literaturist_1[j]},\"{peace_1[j]}\",{econmic}\n")

    for e in range(len(years_2)):
        fd.write(f"{years_2[e]},{Physicist_2[e]},{Chemist_2[e]},{physiologist_2[e]},{literaturist_2[e]},\"{peace_2[e]}\",{econmic_2[e]}\n")
        
import pandas as pd
df = pd.read_csv("List_of_Nobel_laureates.csv")
df 
```
